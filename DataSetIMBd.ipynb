```python
# Célula 3: Download do dataset do IMDB em português
from datasets import load_dataset

print("Baixando o dataset celsowm/imdb-reviews-pt-br...")
# Baixa o dataset público do IMDB
dataset = load_dataset("celsowm/imdb-reviews-pt-br")

print("\nDataset baixado com sucesso! Estrutura dele:")
print(dataset)
```

**4 - Corrigir a coluna, criar os IDs e Salvar o CSV**

```python
# Célula 4: Tratamento dos dados e exportação para o formato do professor
import pandas as pd

print("Convertendo o dataset em tabela...")
# Transforma os dados em uma tabela comum do Pandas
df_original = pd.DataFrame(dataset["train"])

# Seleciona apenas os primeiros 100 documentos (conforme a dica da sua colega)
df_limitado = df_original.head(100).copy()

# Cria uma tabela nova, zerada, para estruturar o padrão do professor
df_final = pd.DataFrame()

# Ação 1: Gera a coluna de ID sequencial iniciando em 1 (1, 2, 3...)
df_final["id"] = range(1, len(df_limitado) + 1)

# Ação 2: Identifica a coluna correta de conteúdo e joga na coluna 'documento'
if "texto" in df_limitado.columns:
    df_final["documento"] = df_limitado["texto"]
else:
    # Se o nome não for exatamente 'texto', ele pega a primeira coluna de dados que achar
    df_final["documento"] = df_limitado.iloc[:, 0]

# Ação 3: Define o caminho exato da pasta do seu Drive
gdrive_path = '/content/drive/MyDrive/Colab Notebooks/SRI/data/documentos.csv'

# Ação 4: Exporta salvando com ponto e vírgula (;) igual ao modelo do professor
df_final.to_csv(gdrive_path, index=False, encoding="utf-8", sep=';')

print("\n=== FLUXO CONCLUÍDO COM SUCESSO ===")
print(f"O arquivo final foi gerado e salvo em: {gdrive_path}")
print(f"Total de registros processados: {len(df_final)}")
```
